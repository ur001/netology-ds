```python
from match_words import *

def print_match_words(word, dict_file_name='sociation_org.txt', min_word_length=3):
    word_matcher = get_word_matcher(word)
    dict_iterator = get_dict_iterator(dict_file_name, min_word_length)    

    for word in filter(word_matcher, dict_iterator):
        print(word)
```

```python
print_match_words('ростелеком', 'sociation_org.txt', 3)
```
